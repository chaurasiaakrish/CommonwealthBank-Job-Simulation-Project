```python
import pandas as pd
import numpy as np 
```

```python
xlsx = pd.ExcelFile("mobile_customers.xlsx")

for sheet in xlsx.sheet_names:
    df = pd.read_excel(xlsx, sheet)
    df.to_csv(f"{sheet}.csv", index=False)
```

```python
df
```

## 1:- Dropping the Important Columns

```python
cols_to_drop = ['username', 'name','gender','address','email','current_location','residence','employer','credit_card_number','credit_card_security_code','job']
to_mask_data=df.drop(columns=cols_to_drop) # storing the unanonimised data in to_mask_data for masking
to_mask_data
```

## 2:- Maksing the leftover Columns

## 2.1:- Masking the Dates

```python
to_mask_data['date_registered']=pd.to_datetime(to_mask_data['date_registered'])
to_mask_data['date_registered']=to_mask_data['date_registered'].dt.year  # to mask the date_registration 
to_mask_data
```

```python
to_mask_data['birthdate']=pd.to_datetime(to_mask_data['birthdate'])
to_mask_data['birthdate']=to_mask_data['birthdate'].dt.year  # to mask the date_registration 
to_mask_data
```

```python
to_mask_data['credit_card_expire'] = pd.to_datetime(to_mask_data['credit_card_expire'],errors='coerce')
to_mask_data['credit_card_expire'] = (to_mask_data['credit_card_expire'].dt.year)
to_mask_data
```
